```python
from sklearn.datasets import load_breast_cancer
import torch
import torch.nn.functional as F
import torch.nn as nn
import matplotlib.pyplot as plt
import time
import numpy as np
from sklearn import datasets
from torch.utils.data import TensorDataset, random_split, DataLoader
from sklearn.preprocessing import StandardScaler
from torchvision import datasets, transforms
import torchvision
```

# Problem 2: Customized Multi-View Dataset

```python
class CustomDataset(torch.utils.data.Dataset):

    def __init__(self):
        self.mnistdata = torchvision.datasets.MNIST(
            root="./data", train=True, download=True
        )

        self.transformations = transforms.Compose(
            [transforms.RandomCrop(17), transforms.RandomRotation(degrees=30)]
        )

    def __len__(self):
        return len(self.mnistdata)

    def __getitem__(self, idx):
        og_image, label = self.mnistdata[idx]
        transform1_image = self.transformations(og_image)
        transform2_image = self.transformations(og_image)

        return label, og_image, transform1_image, transform2_image
```

# Problem 3: Logistic Regression Using Gradient & Newton's Methods

```python
data = load_breast_cancer()
scaler = StandardScaler()
X = torch.tensor(scaler.fit_transform(data["data"]), dtype=torch.float32)
y = torch.tensor(data["target"], dtype=torch.float32)

dataset = TensorDataset(X, y)
```

```python
train_len = int(0.7 * len(dataset))
test_len = len(dataset) - train_len
train_dataset, test_dataset = random_split(dataset, [train_len, test_len])
```

```python
batch_size = 10
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size)
```

## Gradient Descent

```python
# GRADIENT DESCENT
num_iter = 30

model = nn.Linear(in_features=30, out_features=1)  # y = Xw

optimizer = torch.optim.SGD(model.parameters(), lr=0.1)
criterion = nn.BCELoss()

# Train
model.train()
gd_losses, gd_accs, gd_test_times = [], [], []
for i in range(num_iter):
    for x, y in train_loader:
        y = y.unsqueeze(1)
        y_hat = torch.sigmoid(model(x))  # Forward Pass
        loss = criterion(y_hat, y)
        optimizer.zero_grad()
        loss.backward()  # Backward Pass
        # gd_losses.append(loss.item())
        optimizer.step()  # we are using a optimizer now!
    # Evaluate
    t0 = time.perf_counter()
    model.eval()
    with torch.no_grad():
        all_probs, all_targets = [], []
        for x_test, y_test in test_loader:
            y_test = y_test.unsqueeze(1)
            y_test_hat = torch.sigmoid(model(x_test))
            all_probs.append(y_test_hat)
            all_targets.append(y_test)

        probs = torch.cat(all_probs)
        targets = torch.cat(all_targets)

        test_loss = criterion(probs, targets).item()
        preds = (probs >= 0.5).float()
        test_acc = preds.eq(targets).float().mean().item()
    gd_test_times.append(time.perf_counter() - t0)

    gd_losses.append(test_loss)
    gd_accs.append(test_acc)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(gd_accs) + 1), gd_accs)
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Gradient Descent: Accuracy Across Epochs")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(gd_losses) + 1), gd_losses)
plt.xlabel("Epoch")
plt.ylabel("Binary Cross-Entropy Loss")
plt.title("Gradient Descent: Test Loss Across Epochs")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(gd_test_times) + 1), gd_test_times)
plt.xlabel("Epoch")
plt.ylabel("Time")
plt.title("Gradient Descent: Testing Run Timed")
plt.grid(True)
plt.show()
```

## Newton's Method with Exact Expression for Hessian

```python
# NEWTON'S W/ EXACT EXPRESSION FOR HESSIAN

X_train = train_dataset[:][0]  # (N, 30)
y_train = train_dataset[:][1].unsqueeze(1)
X_test = test_dataset[:][0]  # (N_test,  30)
y_test = test_dataset[:][1].unsqueeze(1)  # (N_test,  1)

# Add bias column
Xb_train = torch.cat(
    [X_train, torch.ones(X_train.size(0), 1, dtype=X_train.dtype)], dim=1
)  # (N, 31)
Xb_test = torch.cat(
    [X_test, torch.ones(X_test.size(0), 1, dtype=X_test.dtype)], dim=1
)  # (N_test,  31)
d1 = Xb_train.size(1)

# initialize weights and bias
w = torch.zeros(d1, 1, dtype=Xb_train.dtype)

criterion = nn.BCELoss()

newton_train_losses, newton_test_losses, newton_test_accs, newton_test_times = (
    [],
    [],
    [],
    [],
)
# Train
for i in range(num_iter):

    z = Xb_train @ w
    p = torch.sigmoid(z)

    # gradient
    g = Xb_train.T @ (p - y_train)

    # Hessian !
    s = (p * (1 - p)).squeeze(1)
    XS = Xb_train * s.unsqueeze(1)
    H = Xb_train.T @ XS

    damping = 1e-6
    H = H + damping * torch.eye(d1, dtype=H.dtype)

    # update
    w = w - torch.inverse(H) @ g

    with torch.no_grad():
        train_loss = criterion(p, y_train).item()
        newton_train_losses.append(train_loss)

    t0 = time.perf_counter()
    # evaluate on test
    with torch.no_grad():
        z_te = Xb_test @ w  # logits
        p_te = torch.sigmoid(z_te)  # probs for BCELoss
        test_loss = criterion(p_te, y_test).item()
        preds = (p_te >= 0.5).float()
        test_acc = preds.eq(y_test).float().mean().item()

        newton_test_losses.append(test_loss)
        newton_test_accs.append(test_acc)
    newton_test_times.append(time.perf_counter() - t0)


print(f"Newton final test BCE loss: {newton_test_losses[-1]:.4f}")
print(f"Newton final test accuracy: {newton_test_accs[-1]:.4f}")
```

```python
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(newton_test_losses) + 1), newton_test_losses)
plt.xlabel("Epoch")
plt.ylabel("Binary Cross-Entropy Loss")
plt.title("Newton (Exact Expres. Hessian): Test Loss Across Epochs")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(newton_test_accs) + 1), newton_test_accs)
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Newton (Exact Expres. Hessian): Accuracy Across Epochs")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(newton_test_times) + 1), newton_test_times)
plt.xlabel("Epoch")
plt.ylabel("Time")
plt.title("Newton (Exact Expres. Hessian): Test Run Times Across Epochs")
plt.grid(True)
plt.show()
```

## Newton's Method with Diagonal Approximation for Hessian

```python
# NEWTON'S W/ DIAG APPROXIMATIONFOR HESSIAN

X_train = train_dataset[:][0]  # (N, 30)
y_train = train_dataset[:][1].unsqueeze(1)
X_test = test_dataset[:][0]  # (N_test,  30)
y_test = test_dataset[:][1].unsqueeze(1)  # (N_test,  1)

# Add bias column
Xb_train = torch.cat(
    [X_train, torch.ones(X_train.size(0), 1, dtype=X_train.dtype)], dim=1
)  # (N, 31)
Xb_test = torch.cat(
    [X_test, torch.ones(X_test.size(0), 1, dtype=X_test.dtype)], dim=1
)  # (N_test,  31)
d1 = Xb_train.size(1)

# init weights (including bias)
w = torch.zeros(d1, 1, dtype=Xb_train.dtype)

criterion = nn.BCELoss()

(
    newton_diag_train_losses,
    newton_diag_test_losses,
    newton_diag_test_accs,
    newton_diag_test_times,
) = ([], [], [], [])
# Train
for i in range(num_iter):

    z = Xb_train @ w
    p = torch.sigmoid(z)

    # gradient
    g = Xb_train.T @ (p - y_train)

    # Hessian !
    s = (p * (1 - p)).squeeze(1)
    diag_H = torch.sum((Xb_train**2) * s.unsqueeze(1), dim=0)

    # update
    w = w - (g.squeeze(1) / (diag_H + 1e-6)).unsqueeze(1)

    with torch.no_grad():
        train_loss = criterion(p, y_train).item()
        newton_diag_train_losses.append(train_loss)

    # evaluate on test
    t0 = time.perf_counter()
    with torch.no_grad():
        z_te = Xb_test @ w  # logits
        p_te = torch.sigmoid(z_te)  # probs for BCELoss
        test_loss = criterion(p_te, y_test).item()
        preds = (p_te >= 0.5).float()
        test_acc = preds.eq(y_test).float().mean().item()

        newton_diag_test_losses.append(test_loss)
        newton_diag_test_accs.append(test_acc)
    newton_diag_test_times.append(time.perf_counter() - t0)

print(f"Newton final test BCE loss: {newton_diag_test_losses[-1]:.4f}")
print(f"Newton final test accuracy: {newton_diag_test_accs[-1]:.4f}")
```

```python
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(newton_diag_test_losses) + 1), newton_diag_test_losses)
plt.xlabel("Epoch")
plt.ylabel("Binary Cross-Entropy Loss")
plt.title("Newton: Test Loss Across Epochs")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(newton_diag_test_times) + 1), newton_diag_test_times)
plt.xlabel("Time")
plt.ylabel("Binary Cross-Entropy Loss")
plt.title("Newton: Test Run Time Across Epochs")
plt.grid(True)
plt.show()
```
